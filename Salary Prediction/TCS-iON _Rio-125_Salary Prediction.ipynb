# Splitting the dataset into train and test data

```python
x = dt.drop(['salary'],axis=1)
y = dt['salary']
```

```python
# Scaling
```

```python
x.describe()
```

```python
x.columns
```

```python
sc = preprocessing.StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
x = pd.DataFrame(x,columns=['age','workclass','education','occupation','sex','hours-per-week'])
```

```python
x.describe()
```

# Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42 , test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
model1 = lr.fit(x_train,y_train)
prediction1 = model1.predict(x_test)
```

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score,f1_score
```

```python
accuracy_score(y_test,prediction1)
```

```python
f1_score(y_test,prediction1)
```

# KNN- Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k = []   
neighbors = range(3,18) 

for k in neighbors: 
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier = classifier.fit(x_train,y_train)
    y_pred_kNN = classifier.predict(x_test)
    acc = accuracy_score(y_test,y_pred_kNN)
    metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k values')
plt.ylabel('accuracy')
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors=15)# k value 5 is giving the highest accuracy
classifier = classifier.fit(x_train,y_train)
y_pred_kNN = classifier.predict(x_test)
 
```

```python
accuracy_score(y_test,y_pred_kNN)
```

```python
f1_score(y_test,y_pred_kNN)
```
